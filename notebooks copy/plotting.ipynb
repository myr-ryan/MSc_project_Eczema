```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
```

```python
PROJ_DIR = "/rds/general/user/ym521/home/EczemaNet-DeepLearning-Segmentation-master"
OUTPUT_SWET_DIR = os.path.join(PROJ_DIR, 'output')

# Trained on SWET
SWET_output_dir = os.path.join(PROJ_DIR, 'output_trained_on_SWET/evaluations')
SWET_on_SWET = pd.read_csv(os.path.join(SWET_output_dir, 'skin_test_on_SWET.csv'))
SWET_on_TLA = pd.read_csv(os.path.join(SWET_output_dir, 'skin_test_on_TLA.csv'))
SWET_on_ST = pd.read_csv(os.path.join(SWET_output_dir, 'skin_test_on_S+T.csv'))

# Trained on TLA
TLA_output_dir = os.path.join(PROJ_DIR, 'output_trained_on_TLA/evaluations')
TLA_on_SWET = pd.read_csv(os.path.join(TLA_output_dir, 'skin_test_on_SWET.csv'))
TLA_on_TLA = pd.read_csv(os.path.join(TLA_output_dir, 'skin_test_on_TLA.csv'))
TLA_on_ST = pd.read_csv(os.path.join(TLA_output_dir, 'skin_test_on_ST.csv'))

# # Trained on S+T
ST_output_dir = os.path.join(PROJ_DIR, 'output_trained_on_S+T/evaluations')
ST_on_SWET = pd.read_csv(os.path.join(ST_output_dir, 'skin_test_on_SWET.csv'))
ST_on_TLA = pd.read_csv(os.path.join(ST_output_dir, 'skin_test_on_TLA.csv'))
ST_on_ST = pd.read_csv(os.path.join(ST_output_dir, 'skin_test_on_S+T.csv'))
```

```python
"""
    Trained on SWET
"""
recall_SS, precision_SS, f1_SS = SWET_on_SWET.values[-2, 1:4]
recall_SS_se, precision_SS_se, f1_SS_se = SWET_on_SWET.values[-1, 1:4]
# print(recall_SS, precision_SS, f1_SS)
# print(recall_SS_se, precision_SS_se, f1_SS_se)

recall_ST, precision_ST, f1_ST = SWET_on_TLA.values[-2, 1:4]
recall_ST_se, precision_ST_se, f1_ST_se = SWET_on_TLA.values[-1, 1:4]
# print(recall_ST, precision_ST, f1_ST)
# print(recall_ST_se, precision_ST_se, f1_ST_se)

recall_SST, precision_SST, f1_SST = SWET_on_ST.values[-2, 1:4]
recall_SST_se, precision_SST_se, f1_SST_se = SWET_on_ST.values[-1, 1:4]

"""
    Trained on TLA
"""
recall_TS, precision_TS, f1_TS = TLA_on_SWET.values[-2, 1:4]
recall_TS_se, precision_TS_se, f1_TS_se = TLA_on_SWET.values[-1, 1:4]
# print(recall_TS, precision_TS, f1_TS)
# print(recall_TS_se, precision_TS_se, f1_TS_se)

recall_TT, precision_TT, f1_TT = TLA_on_TLA.values[-2, 1:4]
recall_TT_se, precision_TT_se, f1_TT_se = TLA_on_TLA.values[-1, 1:4]
# print(recall_TT, precision_TT, f1_TT)
# print(recall_TT_se, precision_TT_se, f1_TT_se)

recall_TST, precision_TST, f1_TST = TLA_on_ST.values[-2, 1:4]
recall_TST_se, precision_TST_se, f1_TST_se = TLA_on_ST.values[-1, 1:4]


"""
    Trained on S+T
"""
recall_STS, precision_STS, f1_STS = ST_on_SWET.values[-2, 1:4]
recall_STS_se, precision_STS_se, f1_STS_se = ST_on_SWET.values[-1, 1:4]

recall_STT, precision_STT, f1_STT = ST_on_TLA.values[-2, 1:4]
recall_STT_se, precision_STT_se, f1_STT_se = ST_on_TLA.values[-1, 1:4]


recall_STST, precision_STST, f1_STST = ST_on_ST.values[-2, 1:4]
recall_STST_se, precision_STST_se, f1_STST_se = ST_on_ST.values[-1, 1:4]

```

```python
y = ['Trained on SWET', 'Trained on TLA', 'Trained on S+T']

"""
    Recall plotting
"""
# Tested on SWET
recall_x_S = [recall_SS, recall_TS, recall_STS]
recall_xerr_S = [recall_SS_se, recall_TS_se, recall_STS_se]

# Tested on TLA
recall_x_T = [recall_ST, recall_TT, recall_STT]
recall_xerr_T = [recall_ST_se, recall_TT_se, recall_STT_se]

# Tested on S+T
recall_x_ST = [recall_STST]
recall_xerr_ST = [recall_STST_se]


# For SWET
plt.errorbar(recall_x_S, y, fmt='o', xerr=recall_xerr_S)
# plt.errorbar(recall_x_ST, test_on_ST_y, fmt='o', xerr=recall_xerr_ST)

# plt.xlim([0.994, 1.0])
plt.ylim([0, 1])
plt.xlabel('Tested on SWET')


plt.rcParams.update({'font.size': 15})
# plt.grid(visible=True)
plt.title('Recall - SWET')
# plt.savefig('Recall_SWET.png', bbox_inches='tight')



# # For TLA
# plt.errorbar(recall_x_T, y, fmt='o', xerr=recall_xerr_T)

# plt.xlabel('Tested on TLA')
# plt.rcParams.update({'font.size': 13})
# plt.title('Recall - TLA')
# plt.savefig('Recall_TLA.png', bbox_inches='tight')


# For ST




```
