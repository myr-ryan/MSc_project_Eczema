```python
from shutil import copyfile
from sys import exit
import os
```

```python
DATA_DIR = "/rds/general/user/ym521/home/EczemaNet-DeepLearning-Segmentation-master/data"
source = os.path.join(DATA_DIR, "dataset_TLA/augmented_training_set_TLA/reals")
target = os.path.join(DATA_DIR, "perturbed_dataset_SWET/Blur/severity_1/reals")

# list files
source_dirs = os.listdir(source)
if '.DS_Store' in source_dirs:
    source_dirs.remove('.DS_Store')

# create target folder if not already exists
os.system("mkdir -p " + target)

index = 1


for i in range(len(source_dirs)):
    if 'blur_i5_aug' in source_dirs[i]:
        try:
            copyfile(os.path.join(source, source_dirs[i]), os.path.join(target, source_dirs[i]))
            print('Copying file: ', index)
            index += 1
        except IOError as e:
            print("Unable to copy file. %s" % e)
            exit(1)
        except:
            print("Unexpected error:", sys.exc_info())
            exit(1)

target_dirs = os.listdir(target)

print('Finished copying file, total length is: ', len(target_dirs))
```
